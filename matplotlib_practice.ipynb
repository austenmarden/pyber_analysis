```python
#import dependencies
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis,)

plt.title('$ Per Month')
plt.xlabel('Month')
plt.ylabel('Dollars')
plt.show
```

```python
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="d", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
#Add Axes Labels
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Invert Y axis
plt.gca().invert_yaxis()
plt.title("PyBer Fare by Month")
plt.legend()

```

```python
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
ax.invert_yaxis()
ax.set_title("PyBer Fare by Month")
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
ax.legend()
```

```python
#create scatter plot
plt.scatter(y_axis, x_axis, color="red", label="Chicago")
#label x and y axes
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.gca().invert_yaxis()  #Invert Y axis
plt.legend()  #add legend
plt.title("PyBer Fare by Month")
```

```python
   
plt.scatter(x_axis, y_axis, s=[i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], color="skyblue", label="Boston")
ax.invert_yaxis()
ax.set_title = "PyBer Fares by Month"  # add title to graph
ax.set_xlabel = "Fare($)"
ax.set_ylabel = "Date"
ax.legend()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.5, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, colors=colors, explode=explode_values, labels=x_axis, autopct="%1.1f%%")
plt.show()
```

```python
explode_values = (0, 0, 0.2, 0, 0, 0, 0.4, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
fig, ax=plt.subplots(figsize=(8, 8))
ax.pie(y_axis, labels=x_axis, colors=colors, autopct="%1.1f%%", explode=explode_values, startangle=95, counterclock=False)
plt.show()
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
import numpy as np
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
